```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('simulations_output_all.csv', sep=';', header=None)
df.columns = ['simulation_key', 'missing_steps']
df['controller_name'] = df.simulation_key.str.split('-', expand=True)[0]
df['simulation_index'] = df.simulation_key.str.split('-', expand=True)[1].astype(int)
df.head()
```

```python
df_grouped_zeros = pd.DataFrame(index=df.controller_name.unique(), columns=['zeros_count', 'count'])
for group_name, df_group in df.groupby('controller_name'):
    df_grouped_zeros.loc[group_name, 'zeros_count'] = df_group.missing_steps.isin([0]).sum()
    df_grouped_zeros.loc[group_name, 'count'] = df_group.shape[0]
df_grouped_zeros = df_grouped_zeros.reset_index()
df_grouped_zeros
```

```python
px.bar(df_grouped_zeros, x='index', y='zeros_count', color='index')
```

```python
px.box(df, y='missing_steps', color='controller_name')
```
